#Import the keyed vector


```python
from gensim.models.keyedvectors import KeyedVectors
from sklearn.decomposition import PCA
```

```python
# !gdown "https://drive.google.com/uc?id=1BpfbHu4denceXiv8yfdY3EHgjKIcULku"
embedding_file = './GoogleNews-vectors-negative300.bin.gz'
word_vectors = KeyedVectors.load_word2vec_format(embedding_file, binary=True, limit=500000)
```

#Pre process the data 

##Import the data

```python
import pandas as pd
import numpy as np
import json

df = pd.read_json('dummy.json')
```

```python
df.head()
```

```python
corpus_list = []
for i in range(0, df.shape[0]):
  buffer_str = df['n.documentName'][i].replace("\n", " ") + " " + df['n.pageContent'][i].replace("\n", " ")
  buffer_str = buffer_str.replace("\r", " ")
  corpus_list.append(buffer_str)
```

```python
df['n.pageContent'] = pd.DataFrame(corpus_list)
```

```python
df.head()
```

```python
from bs4 import BeautifulSoup
import re
import string

#removing html code
def remove_html(text):
    soup = BeautifulSoup(text)
    text = soup.get_text()
    return text
df["n.pageContent"] = df["n.pageContent"].apply(remove_html)

#removing urls
def remove_urls(text):
    pattern = re.compile(r'https?://(www\.)?(\w+)(\.\w+)(/\w*)?')
    text = re.sub(pattern, "", text)
    return text
df["n.pageContent"] = df["n.pageContent"].apply(remove_urls)

#removing unicode chars
def remove_unicode_chars(text):
    text = text.encode("ascii", "ignore").decode()
    return text
df["n.pageContent"] = df["n.pageContent"].apply(remove_unicode_chars)

#removing extra spaces
def remove_extra_spaces(text):
    text = re.sub(' +', ' ', text).strip()
    return text
df["n.pageContent"] = df["n.pageContent"].apply(remove_extra_spaces)

#removing punctuations
string.punctuation
def remove_punctuations(text):
    text = re.sub('[%s]' % re.escape(string.punctuation), " ",text)
    return text

df["n.pageContent"] = df["n.pageContent"].apply(remove_punctuations)
```

```python
df['n.pageContent'][0]
```

```python
import spacy

nlp = spacy.load('en_core_web_sm')
unwanted_pipes = ['ner', 'parser']

# Further remove stop words and take the lemma instead of token text.
def spacy_tokenizer(doc):
  with nlp.disable_pipes(*unwanted_pipes):
    return [t.lemma_ for t in nlp(doc) if \
            ##Try with and without punctuation, as punctuation holds some sentiment/emotion value
            not t.is_punct and \
            not t.is_space and \
            not t.is_stop]

df["tokenized.pageContent"] = df["n.pageContent"].apply(spacy_tokenizer)
```

```python
df.head()
```

#Compare and get rankings

```python
query = "when do i get my security deposit back?"
query = spacy_tokenizer(query)
print(query)
```

```python
new_scores = []

for i in range(0, df.shape[0]):
  candidate_corpus = df['tokenized.pageContent'][i]
  similarity = word_vectors.n_similarity(query, candidate_corpus)
  new_scores.append(similarity)

df['scores'] = pd.DataFrame(new_scores)
```

```python
new_df = df.sort_values('scores', ascending=False)
```
